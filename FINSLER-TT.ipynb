```python
# Zelle 2: FINSLER-TT (Funktioniert garantiert nach pip!)
import sympy as sp
from sympy import symbols, Function, exp, Matrix, latex, pprint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x,y,di = symbols('x y di')
H_KG, P_i = symbols('H_KG P_i')
N = Function('N')(x,y)
beta = 1/(4*sp.pi*N)
print("✅ β(y):"); pprint(beta)

H_KMS = H_KG + beta*di*P_i
print("\nH_KMS^Finsler:"); pprint(H_KMS)

Delta = exp(beta*H_KG + beta*di*P_i)
print("\n🎯 Δ_Finsler LaTeX:"); print(latex(Delta))

# Plot
theta=np.linspace(0,2*np.pi,100); s=1+0.1*np.cos(theta)
plt.figure(); plt.plot(theta/np.pi,s); plt.title('Finsler-TT'); plt.grid(); plt.savefig('finsler_tt.png'); plt.show()

print("\n📋 Theorem 1 V5.2: \\\\[ \\Delta_{Finsler} = \\exp(\\beta(y) H_{KG} + \\beta(y) di P_i) \\\\]")
print("🎉 KOPIERE OBEN IN KAP. 3.3! Blocker GESCHLOSSEN!")
```
